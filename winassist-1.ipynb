```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
winassit = pd.read_csv('winassist_responses.csv')
winassit.head()
```

```python
winassit.rename(columns={'Timestamp':'Timestamp',
                         'Vilken produkt använder du hos Winassist? Det går att välja flera alternativ.':'använder_product',
                        'Upplever du att Winassist uppdaterar dig med information som nyheter, uppdateringar kring produkten och erbjudanden?':
                        'uppdaterar_information',
                        'Har du använt Winassists teknikerapp?':'använt_teknikerapp',
                        'Vad för information tycker du är viktig för dig att ta del av och skulle vilja se mer av? Det går att välja flera alternativ.'
                        :'viktig_information',
                        'Om du fick välja, hur skulle den informationen du valde i fråga 3 kommuniceras? Det går att välja flera alternativ.':
                        'kommunicerasty_typ'}, inplace=True)
winassit.head()
```

```python
plt.figure(figsize=(10,6))
sns.set_context('notebook')
fig = sns.countplot(x='uppdaterar_information', data=winassit)
fig.set_title('Upplever du att Winassist uppdaterar dig med information som nyheter,uppdateringar kring produkten och erbjudanden?',y=1.03)
fig.set(xlabel='Uppdaterar med information som nyheter', ylabel='Number of counts')
plt.savefig('1.png',transparent=True)
```

```python
plt.figure(figsize=(10,6))
sns.set_context('notebook')
fig = sns.countplot(x='använt_teknikerapp', data=winassit)
fig.set_title('Har du använt Winassists teknikerapp?', y=1.03)
fig.set(xlabel='Använt Winassists teknikerapp', ylabel='Number of counts')
plt.savefig('2.png',transparent=True)
```

```python
Winassist_Verkstad = winassit['använder_product'].str.contains('Winassist Verkstad').sum()
Winassist_Fordon = winassit['använder_product'].str.contains('Winassist Fordon').sum()
Winassist_Kassa = winassit['använder_product'].str.contains('Winassist Kassa').sum()
Winassist_Ekonomi = winassit['använder_product'].str.contains('Winassist Ekonomi').sum()
Winassist_Marin = winassit['använder_product'].str.contains('Winassist Marin').sum()
list = [Winassist_Verkstad,Winassist_Fordon,Winassist_Kassa,Winassist_Ekonomi,Winassist_Marin]

plt.figure(figsize=(15,8))
sns.set_style('white')
sns.set_context('talk')
plt.bar(range(5), list)
plt.title('Vilken produkt använder du hos Winassist?', y=1.03)
ax = plt.subplot()
ax.set_xticks([0,1,2,3,4])
ax.set_xticklabels(['Winassist_Verkstad','Winassist_Fordon','Winassist_Kassa','Winassist_Ekonomi','Winassist_Marin'],rotation=15)
plt.savefig('3.png',transparent=True)
```

```python
Email = winassit['kommunicerasty_typ'].str.contains('Email').sum()
Nyhetsflöde = winassit['kommunicerasty_typ'].str.contains('Nyhetsflöde i produkten').sum()
Startsida = winassit['kommunicerasty_typ'].str.contains('Startsida i produkten').sum()
hemsida = winassit['kommunicerasty_typ'].str.contains('Via Winassist hemsida').sum()
Notis = winassit['kommunicerasty_typ'].str.contains('Notis i produkten').sum()
Telefonsamtal = winassit['kommunicerasty_typ'].str.contains('Telefonsamtal').sum()
Utbildning_på_plats = winassit['kommunicerasty_typ'].str.contains('Utbildning på plats').sum()
Visste_inte = winassit['kommunicerasty_typ'].str.contains('Visste inte om att det fanns en teknikerapp').sum()

list1 = [Email,Nyhetsflöde,Startsida,hemsida,Notis,Telefonsamtal,Utbildning_på_plats,Visste_inte]
print(list1)

plt.figure(figsize=(15,8))
sns.set_style('white')
sns.set_context('talk')
plt.bar(range(8), list1)
plt.title('Om du fick välja, hur skulle den informationen du valde i fråga 3 kommuniceras? ', y=1.03)
ax = plt.subplot()
ax.set_xticks([0,1,2,3,4,5,6,7])
ax.set_xticklabels(['Email','Nyhetsflöde','Startsida','hemsida','Notis','Telefonsamtal','Utbildning_på_plats','Visste_inte'],rotation=15)
plt.savefig('4.png',transparent=True)
```
